```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
import datetime
def dateparse (time_in_secs):    
    return datetime.datetime.fromtimestamp(float(time_in_secs))
```

```python
btc = pd.read_csv('../falconx/XBTUSD_60.csv', parse_dates=['Timestamp'], date_parser=dateparse,  names=['Timestamp', 'Open', 'High', 'Low', 'Close', 'Volume', 'dunno']) 
eth = pd.read_csv('../falconx/ETHUSD_60.csv', parse_dates=['Timestamp'], date_parser=dateparse,  names=['Timestamp', 'Open', 'High', 'Low', 'Close', 'Volume', 'dunno']) 
stj = pd.read_csv('../falconx/STORJUSD_60.csv', parse_dates=['Timestamp'], date_parser=dateparse,  names=['Timestamp', 'Open', 'High', 'Low', 'Close', 'Volume', 'dunno']) 
```

```python
plt.figure(figsize=(30,10))
plt.plot(btc['Timestamp'], btc['Close'], label="BTC in USD")
plt.plot(eth['Timestamp'], eth['Close'], label="ETH in USD")
plt.plot(stj['Timestamp'], stj['Close'], label="STORJ in USD")
plt.legend()
plt.show()
```

## Calculating percent change and log of returns

```python
btc['pct_change'] = btc['Open'].pct_change()
btc['log_return'] = np.log1p(btc['Open'].pct_change())

eth['pct_change'] = eth['Open'].pct_change()
eth['log_return'] = np.log1p(eth['Open'].pct_change())

stj['pct_change'] =stj['Open'].pct_change()
stj['log_return'] = np.log1p(stj['Open'].pct_change())
```

```python
fig, axis = plt.subplots(3, figsize=(50,15), sharex=True)
fig.suptitle('Log returns comparison of selected Crypto')
axis[0].plot(btc['Timestamp'], btc['log_return'], label="log return of BTC", )
axis[0].legend()

axis[1].plot(eth['Timestamp'], eth['log_return'], label="log return of ETH")
axis[1].legend()

axis[2].plot(stj['Timestamp'], stj['log_return'], label="log return of STORJ")
axis[2].legend()
```

```python
fig, axis = plt.subplots(3, figsize=(50,15), sharex=True)
fig.suptitle('Log returns comparison of selected Crypto')
axis[0].plot(btc['Timestamp'], btc['pct_change'], label="log return of BTC", )
axis[0].legend()

axis[1].plot(eth['Timestamp'], eth['pct_change'], label="log return of ETH")
axis[1].legend()

axis[2].plot(stj['Timestamp'], stj['pct_change'], label="log return of STORJ")
axis[2].legend()
```
